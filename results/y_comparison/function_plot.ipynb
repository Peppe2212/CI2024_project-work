```python
def compare_y_pred_target(func, X, target, problem_number):
    """
    Compare the predicted values from a function with the target values and save the results.

    Parameters:
    - func: The function used to generate predictions.
    - X: The feature matrix (numpy array).
    - target: The target values (numpy array).
    - problem_number: Identifier for the dataset/problem.
    """
    # Evaluate the function on the dataset
    y_pred = evaluate_function_on_dataset(func, X)

    # Create a DataFrame with the target, predicted values, and the absolute difference
    df = pd.DataFrame({
        "y_true": target,
        "y_pred": y_pred,
        "y_diff": np.abs(target - y_pred)
    })

    # Ensure the directory exists
    directory = "./"
    os.makedirs(directory, exist_ok=True)

    # Save the DataFrame to a CSV file
    file_path = os.path.join(directory, f"comparing_result_{problem_number}.csv")
    df.to_csv(file_path, index=False)

    print(f"Comparison results saved to: {file_path}")

def load_npz_dataset(filepath: str):
    """
    Load a dataset from an .npz file.

    Parameters:
    - filepath: Path to the .npz file.

    Returns:
    - X: Feature matrix.
    - y: Target values.
    """
    data = np.load(filepath)
    X = data['x'].T
    y = data['y']  
    return X, y

def evaluate_function_on_dataset(func, X):
    """
    Evaluate a function on a dataset.

    Parameters:
    - func: The function to evaluate.
    - X: The feature matrix.

    Returns:
    - Predictions as a numpy array.
    """
    return np.array([func(x) for x in X])

filepath = "/Users/gabriele/Desktop/Magistrale/computational intelligence/Code/CI_project/dataset_project/problem_1.npz"

# Load dataset
X, y = load_npz_dataset(filepath)

# Example: Compare predictions from f1
problem_number = 1
compare_y_pred_target(f1, X, y, problem_number)
```
