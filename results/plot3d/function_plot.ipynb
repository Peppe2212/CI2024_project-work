```python
def plot_results_1D(best_program, features, target):
    import numpy as np
    import plotly.graph_objects as go


    X = features[:, 0]

    
    x_grid = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

   
    y_pred_flat = np.array([best_program(np.array([x])) for x in x_grid.ravel()])
    y_pred_grid = y_pred_flat.reshape(x_grid.shape)

 
    fig = go.Figure()

   
    fig.add_trace(go.Scatter(
        x=X,
        y=target,
        mode='markers',
        marker=dict(size=4, color='blue'),
        name="Real Data"
    ))


    fig.add_trace(go.Scatter(
        x=x_grid.ravel(),
        y=y_pred_grid.ravel(),
        mode='lines',
        line=dict(color='red'),
        name="Approximation"
    ))

  
    fig.update_layout(
        xaxis_title='x',
        yaxis_title='y',
        title="Comparison between real points and approximated function",
        autosize=True
    )


    fig.show()
def plot_results_3D(best_program, features, target):
    import numpy as np
    import plotly.graph_objects as go
    import plotly.io as pio


    X1 = features[:, 0]
    X2 = features[:, 1]


    x1_grid = np.linspace(X1.min(), X1.max(), 50)
    x2_grid = np.linspace(X2.min(), X2.max(), 50)
    x1_mesh, x2_mesh = np.meshgrid(x1_grid, x2_grid)

 
    data_points = np.array([x1_mesh.ravel(), x2_mesh.ravel()]).T


    y_pred_flat = np.array([best_program(point) for point in data_points])
    y_pred_mesh = y_pred_flat.reshape(x1_mesh.shape)


    fig = go.Figure()


    fig.add_trace(go.Scatter3d(
        x=X1,
        y=X2,
        z=target,
        mode='markers',
        marker=dict(size=4, color='blue'),
        name="Real Data"  # Label for real data points
    ))

    # Add the predicted surface
    fig.add_trace(go.Surface(
        z=y_pred_mesh,
        x=x1_grid,
        y=x2_grid,
        colorscale='Reds',
        opacity=0.7,
        name="Approximation"  # Label for the approximated function
    ))

    # Configure the layout
    fig.update_layout(
        scene=dict(
            xaxis_title='x1',
            yaxis_title='x2',
            zaxis_title='y'
        ),
        title="Comparing the Real and Predicted Data",
        autosize=True,
        legend=dict(
            title="Legend",
            itemsizing="constant"
        )
    )

    # Save the plot as an HTML file
    file_path = "plot3d_7.html"
    pio.write_html(fig, file=file_path, auto_open=True)

    # Print the link to the HTML file
    print(f"The 3D plot has been saved here: {file_path}")

    # Show the plot
    fig.show()

    return file_path
```
